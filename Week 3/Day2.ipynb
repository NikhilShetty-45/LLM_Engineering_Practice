## Hugging face Pipeline

2 API levels of Hugging face

1) Pipelines : Higher level APis to carry out std tasks incredibly quickly

Pipelines are for simple out of the box interface tasks : Sentiment Analysis, Classifier, Named Entity Recognition, Question Answering, Summarizing and Translation

2) Tokenizers & Models : Lower Level APis to provide the most power and control

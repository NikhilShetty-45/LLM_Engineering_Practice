## Introducing The Tokenizer 

Maps between Text adn Tokens for a particular model

• Translates between Text, Tokens and Token IDs with encode() & decode() methods

• Contains a vocab that includes special tokens to signal information to the LLm, like start of prompt

Ex: 

1. Llama 3.1 

2. Phi 4

3. Deepseek 3.1

4. Qwen 2.2 Coder

```python
from dotenv import load_dotenv
from IPython.display import display, Markdown
from openai import OpenAI
```

```python
load_dotenv(override=True)

model_name = "gpt-5-mini"
```

```python
openai= OpenAI()
```

```python
messi_system_prompt = """
You are Messi Fan, a chatbot who is a die-hard Lionel Messi supporter. 
You always consider Messi the GOAT and declare he is the best footballer in the world. 
If anyone claims another player is better than Messi by providig some stats, you immediately give counter Messi stat arguement against them which is not used previously. 
Your replies must always be short, limited to 1–2 sentences.
"""

ronaldo_system_prompt = """
You are Ronaldo Fan, a chatbot who is a die-hard Cristano Ronaldo supporter. 
You always consider Ronaldo the GOAT and declare he is the best footballer in the world. 
If anyone claims another player is better than Ronaldo by providig some stats, you immediately give counter Ronaldo stat arguement against them which is not used previously.
Your replies must always be short, limited to 1–2 sentences.
"""

neymar_system_prompt = """
You are Neymar Fan, a chatbot who is a die-hard neymar JR supporter. 
You always consider neymar the GOAT and declare he is the best footballer in the world. 
If anyone claims another player is better than Neymar by providig some stats, you immediately give counter Neymar stat arguement against them which is not used previously.
Your replies must always be short, limited to 1–2 sentences.
"""
```

```python
conversation = []
```

```python
def call_fan(player_name, system_prompt):
    user_prompt = f"""
        You are {player_name} fan, in conversation with the other football fans.
        The conversation so far is as follows:
        {conversation}
        Now respond with what you would like to say next, as {player_name} fan.
    """ 
    response = openai.chat.completions.create(
        model = model_name,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    return response.choices[0].message.content
```

```python
conversation.append("Messi is the Goat\n")
conversation.append("No, Ronaldo is the Goat\n")
conversation.append("Neither, Neymar is the Goat\n")

display(Markdown(f"### Messi Fan:\n {conversation[0]}\n"))
display(Markdown(f"### Ronaldo Fan:\n{conversation[1]}\n"))
display(Markdown(f"### Neymar Fan:\n {conversation[2]} \n"))

for i in range(3):
    messi_next = f"### Messi Fan:\n {call_fan("Lionel Messi",messi_system_prompt)}\n"
    display(Markdown(messi_next))
    conversation.append(messi_next)
    
    ronaldo_next = f"### Ronaldo Fan:\n {call_fan("Cristiano Ronaldo", ronaldo_system_prompt)}\n"
    display(Markdown(ronaldo_next))
    conversation.append(ronaldo_next)

    neymar_next = f"### Neymar Fan:\n {call_fan("Neymar Jr", neymar_system_prompt)}\n"
    display(Markdown(neymar_next))
    conversation.append(neymar_next)
```

To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2:1b'
```

```python
# set up environment
load_dotenv(override=True)

OLLAMA_BASE_URL = "http://localhost:11434/v1"

openai = OpenAI()
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
system_prompt = """
 You are professional technical trainer
 Who take technival question from the user and answer them in a clearly and in simple langauge
"""

user_prompt = "Here is my technical question : "
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
def get_openai_response(question):
    response = openai.chat.completions.create(
        model = MODEL_GPT,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt + question}
        ]
    )
    return response.choices[0].message.content
```

```python
# Get Llama 3.2 to answer
def get_ollama_response(question):
    response = ollama.chat.completions.create(
        model = MODEL_LLAMA,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt + question}
        ]
    )
    return response.choices[0].message.content
```

```python
display(Markdown(get_openai_response(question)))
```

```python
display(Markdown(get_ollama_response(question)))
```
